# Tuple


```python
a = 4, 6, 7, 78, 3, 4, 3, 5, (3, 5, 64)
print(a)
print(type(a))
```

#### Accessing Elements


```python
# Indexing and Slicing
t1 = a[1]
t2 = a[4:7]
t3 = a[-1][-1]

# a[3]=34 ->  'tuple' object does not support item assignment

print(f"Variable value is: {t3}")
```

##### Tuple Operations


```python
a = (5, 7, 39, 345, 463)
b = (43, 5, 3, 42, 2)

c= a * 3
print(c)

a = a + b
print(a)

print(435 in a)
```

##### Tuple Functions


```python
print(a.count(42))
print(len(a))
print(min(a))
print(max(a))
print(sorted(a))
```
